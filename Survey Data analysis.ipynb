```python
#missing data analysis
all_data_na = (df.isnull().sum() / len(df)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)[:30]
missing_data = pd.DataFrame({'Missing Ratio' :all_data_na})
missing_data
```

```python
#visualize missing data
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
f, ax = plt.subplots(figsize=(10, 8))
plt.xticks(rotation='90')
sns.barplot(x=all_data_na.index, y=all_data_na)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
```

```python
df['[Are you keen on moving closer to your work place?]'].fillna('no',inplace = True)
```

```python
#drop unnecessary data
df.drop('[Have you ever worked before?]', axis = 1, inplace = True)
df.drop('[How much would you like to get monthly - refuse to answer]', axis = 1,  inplace = True)
df.drop('[What is your current or previous monthly net salary - refuse to answer]', axis = 1,  inplace = True)
df.drop('[Are you looking for a job?]', axis = 1,  inplace = True)
df.drop('[Q13-C22 - Do you work currently? - Answer]', axis = 1,  inplace = True)
```

```python
df.drop('Unnamed: 0', axis = 1,  inplace = True)
df.drop('id', axis = 1,  inplace = True)
```

```python
#task 3
#encode willingness to work to 1 and 0's 
def encode(a):
    if a == 'yes':
        a = 1
    else:
        a = 0
    return a

df['[Are you keen on moving closer to your work place?]'] = df['[Are you keen on moving closer to your work place?]'].apply(encode)
```

```python
#create categories out of columns
df2 = pd.DataFrame()
df2['[D1 City size]'] =df['[D1 City size]'].astype('category').cat.codes
df2['[D2 Gender]'] =df['[D2 Gender]'].astype('category').cat.codes
df2['[D3 Age]'] =df['[D3 Age]'].astype('category').cat.codes
df2['[D4 Education]'] =df['[D4 Education]'].astype('category').cat.codes
df2['[Are you keen on moving closer to your work place?]'] = df['[Are you keen on moving closer to your work place?]']


df2[df2.columns[:]].corr()['[Are you keen on moving closer to your work place?]'][:]
```

```python
#machine learning approach. it will automaticall show most important features
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
# load data

# split data into X and y
X = df2.drop('[Are you keen on moving closer to your work place?]', axis =1)
y = df2['[Are you keen on moving closer to your work place?]']


cols = []
for col in X.columns:
    col = col.replace('[','')
    col = col.replace(']','')
    cols.append(col)
X.columns = cols

# fit model no training data
model = XGBClassifier()
model.fit(X, y)
print(model.feature_importances_)
# plot
plt.bar(range(len(model.feature_importances_)), model.feature_importances_)
features = ['[D1 City size]','[D2 Gender]','[D3 Age]','[D4 Education]']
plt.xticks( np.arange(len(model.feature_importances_)),features)
plt.show()
```

```python
wm = ['[D1 City size]','[D2 Gender]','[D3 Age]','[D4 Education]','[Are you keen on moving closer to your work place?]']
dftask1_v2 = pd.get_dummies(df[wm])
dftask1_v2.head()
```

```python
corrmat = dftask1_v2.corr()
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, '[Are you keen on moving closer to your work place?]')['[Are you keen on moving closer to your work place?]'].index
cm = np.corrcoef(dftask1_v2[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
X1 = dftask1_v2.drop('[Are you keen on moving closer to your work place?]', axis =1)
y1 = dftask1_v2['[Are you keen on moving closer to your work place?]']


cols = []
for col in X1.columns:
    col = col.replace('[','')
    col = col.replace(']','')
    cols.append(col)
X1.columns = cols

# fit model to training data
model1 = XGBClassifier()
model1.fit(X1, y1)
print(model1.feature_importances_)
# plot
features = X1.columns
plt.bar(range(len(model1.feature_importances_)), model1.feature_importances_)

plt.show()
```
